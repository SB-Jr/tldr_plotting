# Using Pandas Dataframe

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

We will create a dataset containing some fictonal data about the no of immigrations from India and China to Canada from 1980 to 1984

```python
immigrants = pd.DataFrame({
    'India':[8880, 8670, 8147, 7338, 5704],
    'China':[5123, 6682, 3308, 1863, 1527]},
    index=[1980, 1981, 1982, 1983, 1984]
)
```

```python
immigrants
```

## Bar Plot

### Simple

```python
immigrants.plot(kind='bar')
plt.show()
```

### Basic Detailing

```python
immigrants.plot(kind='bar')
plt.grid() # shows a grid
plt.title('Immigration to Canada')
plt.show()
```

```python
sns.set_style('ticks')
immigrants.plot(kind='bar')
plt.grid() # shows a grid
plt.title('Immigration to Canada')
plt.show()
```

## Line Graph

```python
sns.set_style('dark')
immigrants.plot(kind='line')
plt.grid()
plt.title('Immigrations to Canada')
plt.show()
```

## Histogram

```python
sns.set_style('white')
immigrants.plot(kind='hist')
plt.grid()
plt.title('Immigrants to Canada')
plt.plot()
```

## Line Plot

```python
sns.set_style('ticks')
immigrants.plot(kind='line')
plt.xlabel('Years')
plt.ylabel('Immigrations')
plt.title('Immigrations to Canada')
plt.show()
```

```python
sns.set_style('ticks')
immigrants.plot(figsize=(10,10))
plt.xlabel('Years')
plt.ylabel('Immigrations')
plt.title('Immigrations to Canada')
plt.show()
```

## Area Plot

```python
fon_settings={
    'fontsize': 20,
    'verticalalignment': 'baseline',
    'color': 'Red'
}
```

```python
sns.set_style('ticks')
immigrants.plot(kind='area', figsize=(10,5))
plt.xlabel('Years', fontdict=fon_settings)
plt.ylabel('Immigrations', fontdict=fon_settings)
plt.title('Immigrations to Canada')
plt.show()
```

## Unstacked Area Graph

```python
sns.set_style('ticks')
immigrants.plot(kind='area', figsize=(10, 5), stacked=False)
plt.xlabel('Years')
plt.ylabel('Immigrations')
plt.title('Immigrations to Canada')
plt.show()
```
