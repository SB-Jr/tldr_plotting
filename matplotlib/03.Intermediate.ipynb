# Multi Graph Settings

## Axes

Axes are the objects which contain the graph that we want to visualize. A figure can contain multiple axes objects but 1 axes object can only contain 1 figure.


Each Axes object contain 2 or 3 Axis objects based on whether it is a 2D graph or a 3D graph.


We can add axes to a figure by using the `fig_object.add_axes([left, bottom, width, height])` where the values are respect to the actual figure object and can be fractional.

When we need to plot a graph in a specific axes, we can use the `axes_object.plot()` or `.imshow()` methods

## 2 Graphs

We will use the same data used in the 2nd notebook.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
immigrants = pd.DataFrame({
    'India':[8880, 8670, 8147, 7338, 5704],
    'China':[5123, 6682, 3308, 1863, 1527]},
    index=[1980, 1981, 1982, 1983, 1984]
)
```

```python
# Step1.A: Create a figure
fig = plt.figure(figsize=(10, 5))
# Step1.B: set the style
sns.set_style('ticks')

# Step2: Add subplots to the figure
ax0 = fig.add_subplot(121) # 1 row, 2 cols, 1st item
ax1 = fig.add_subplot(122) # 1 row, 2 cols, 2nd item

# Step3: Define the Subplots
immigrants.plot(kind='area', alpha=0.4, ax=ax0)
ax0.set_title('Area Stacked')
ax0.set_xlabel('Years')
ax0.set_ylabel('Immigrants')

immigrants.plot(kind='area', stacked=False, alpha=0.4, ax=ax1)
ax1.set_title('Area Unstacked')
ax1.set_xlabel('Years')
ax1.set_ylabel('Immigrants')

# Step4: Show the plot
plt.show()
```

## Playing with Colors

```python
import matplotlib as mpl
```

```python
color_list = []
for name, hex in mpl.colors.cnames.items():
    color_list.append((name, hex))
len(color_list)
```

From the above code we can see that the Matplotlib has about 148 colors to choose from when we need to style our graphs

```python
import numpy as np
```

```python
fig = plt.figure(figsize=(10,10))
for i in range(6):
    subplot_index = 231+i
    ax = fig.add_subplot(subplot_index)
    x = np.random.randint(0, len(color_list))
    ax.set_title(color_list[x][0])
    immigrants.plot(kind='bar', color=color_list[x][0], ax=ax)
plt.show()
```

## Overlapping Plots

```python
x = np.arange(0, np.math.pi*2, 0.05)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax2 = fig.add_axes([0.55, 0.55, 0.3, 0.3])
ax1.plot(x, y1, 'b')
ax1.set_title('sine')
ax2.plot(x, y2, 'r')
ax2.set_title('cosine')
plt.show()
```
