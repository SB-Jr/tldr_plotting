# MatplotLib.Pyplot Examples

## This notebook comprises of beginner to basic expert level examples and syntaxes on how to use matplotlib.pyplot  for everyday use.

A Matplotlib figure can be categorized into several parts as below:
<ol>
    <li><b>Figure</b>: It is a whole figure which may contain one or more than one axes (plots). You can think of a Figure as a canvas which contains plots.</li>
    <li><b>Axes</b>: It is what we generally think of as a plot. A Figure can contain many Axes. It contains two or three (in the case of 3D) Axis objects. Each Axes has a title, an x-label and a y-label.</li>
    <li><b>Axis</b>: They are the number line like objects and take care of generating the graph limits.</li>
    <li><b>Artist</b>: Everything which one can see on the figure is an artist like Text objects, Line2D objects, collection objects. Most Artists are tied to Axes.</li>

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Basic usage

```python
plt.plot([1,2,3,4],[4,5,6,7])
plt.show()
```

## Adding basic info to a graph

```python
plt.plot([1,2,3,4],[5,7,8,9])
plt.title('A simple graph')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot([1,2,3,4],[5,7,8,9])
plt.show()
```

## Adding multiple plots in the same graph

```python
x1 = np.arange(1,10)
y1 = np.arange(10,100,10)

x2 = np.arange(1,10)
y2 = x2**3    

plt.plot(x1,y1,'b-',x2,y2,'g-')
plt.title('2 plots in the same graph')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.show()
```

## Adding subplots in the same plot object

We can use subplot() method to add more than one plots in one figure. In the image below, we used this method to separate two graphs which we plotted on the same axes in the previous example. The subplot() method takes three arguments: they are <b>n-rows</b>, <b>n-cols</b> and <b>index</b>. They indicate the number of rows, number of columns and the index number of the sub-plot.

```python
x1 = np.arange(1,10)
y1 = np.arange(10,100,10)

x2 = np.arange(1,10)
y2 = x2**3    

plt.subplot(1,2,1)
plt.plot(x1,y1,'b-')
plt.title('1st plot')
plt.xlabel('1st X-Label')
plt.ylabel('1st Y-Label')

plt.subplot(1,2,2)
plt.plot(x2,y2,'g-')
plt.title('2nd plot')
plt.xlabel('2nd X-Label')
plt.ylabel('2nd Y-Label')

plt.suptitle('2 plots in 1 graph')
plt.show()
```

```python
x1 = np.arange(1,10)
y1 = np.arange(10,100,10)

x2 = np.arange(1,10)
y2 = x2**3    

plt.subplot(2,1,1)
plt.plot(x1,y1,'b-')
plt.title('1st plot')
plt.xlabel('1st X-Label')
plt.ylabel('1st Y-Label')

plt.subplot(2,1,2)
plt.plot(x2,y2,'g-')
plt.title('2nd plot')
plt.xlabel('2nd X-Label')
plt.ylabel('2nd Y-Label')

plt.suptitle('2 plots in 1 graph')
plt.show()
```

A more convenient way is to use <b>subpltots()</b> method. Notice the difference of <b>‘s’</b> in both the methods. This method takes two arguments </b>n-rows</b> and </b>n-cols</b> as number of rows and number of columns respectively. This method <b>creates two objects:figure and axes<b> which we store in variables fig and ax which can be used to change the figure and axes level attributes respectively.

```python
x1 = np.arange(1,10)
y1 = np.arange(10,100,10)

x2 = np.arange(1,10)
y2 = x2**3

figs, axs = plt.subplots(nrows=2, ncols=1)

axs[0].plot(x1,y1,'g-')
axs[0].set_title('Line graph')

axs[1].plot(x2,y2,'b-')
axs[1].set_title('Cube graph')

plt.show()
```

## Adding different kinds of graphs

### Bar graph

#### Vertical Bar Graph

```python
divisions = ['Div-A','Div-B','Div-C','Div-D']
div_value = [80,100,70,60]

plt.bar(divisions,div_value,color='blue')
plt.title('A Bar graph')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.show()
```

#### Horizontal Bar Graph
To make horizontal bar graphs use method barh() Also we can pass an argument (with its value)xerr oryerr (in case of the above vertical bar graphs) to depict the variance in our data.

```python
divisions = ['Div-A','Div-B','Div-C','Div-D']
div_value = [80,100,70,60]
variance = [10,20,30,40]

plt.barh(divisions,div_value,color='green',xerr=variance)
plt.title('A Bar graph')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.show()
```

#### Multiple Stacked Bar Graph
To create horizontally stacked bar graphs we use the bar() method twice and pass the arguments where we mention the index and width of our bar graphs in order to horizontally stack them together. Also, notice the use of two other methods legend() which is used to show the legend of the graph and xticks() to label our x-axis based on the position of our bars.

We need to use index and specify the width to stack it side by side

```python
division = ['Div-A','Div-B','Div-C','Div-D']
div_stack1_value = [80,100,70,60]
div_stack2_value = [100,90,80,70]

index = np.arange(4)
width = 0.3

plt.bar(index,div_stack1_value,width,color='green',label='Stack 1 Value')
plt.bar(index + width, div_stack2_value,width,color='blue',label='Stack 2 Value')

plt.title('Multiple Stacked Bar graph')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.xticks(index + width/2, division)  # separates the bars from each other

plt.legend(loc='best')
plt.show()
```
